### Pandas Series

```python
#Pandas series
g7 = pd.Series([3.05, 6.95, 5.0, 6.7])
```

This series is a single column of a dataframe

```python
g7
```

```python
g7.name = "hehe"
g7
```

```python
g7.dtype
```

```python
#selecting third value of third row
g7.iloc[2]
```

```python
#custom index
g7.index = ["lol", "hehe", "kuku", "hihi"]
```

```python
g7
```

```python
#indexing via the custom indices
g7["hehe"]
```

```python
#selecting multiple values via indexing
g7[["lol", "hehe"]]   #upper limit is included
```

```python
#selecting multiple values via iloc
g7.iloc[[0,1]]
```

```python
g7.mean()
```

```python
g7 > 5
```

```python
g7[g7 >5]
```

```python
g7[g7 > g7.mean()]
```

```python
df_cars.info()
```

```python
#Selecting rows by index by using the loc attribute
df_cars.iloc[3:5]
```

```python
#selecting entire column
df_cars["COMPACTNESS"]
```

```python
#using iloc to select rows and columns
df_cars.iloc[1:5, 0:3]
```

```python
#conditional selection
df_cars.loc[df_cars["CIRCULARITY"] > 45]
```

```python
#more complex conditional selection
df_cars.loc[df_cars["COMPACTNESS"] > 50, ["COMPACTNESS", "CIRCULARITY"],]
```

```python
circularity = df_cars["CIRCULARITY"]
```

```python
circularity.mean()
```

```python
#Access a single value for a row/column label pair
df_cars.at[1,"COMPACTNESS"]
```

```python
#plotting two variables into one plot
df_cars.loc[:, ["CIRCULARITY" , "COMPACTNESS"]].plot()
```
